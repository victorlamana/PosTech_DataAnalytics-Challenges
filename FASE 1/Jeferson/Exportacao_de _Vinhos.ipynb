```python
def plota_kg(dados, figsize=(10,7)): # uma função em python
  axis = dados.T.plot(kind='line',figsize=figsize)
  axis.set_ylabel("Milhões de litros de Vinho")
  axis.set_xlabel("Ano")
  plt.title('Exportações de vinho nacional entre 2006 a 2021')
  plt.grid()
  axis.set_xticks(range(15))

```

```python
df_vinho_kg_ordenado_drop = df_vinho_kg_ordenado_drop / 1_000_000
```

```python
plota_kg(df_vinho_kg_ordenado_drop.head(15))
plt.savefig('grafico_com_russia.png')
```

```python
df_vinho_kg_ordenado_drop_russia = df_vinho_kg_ordenado_drop.drop('Rússia', axis=0)
```

```python
plota_kg(df_vinho_kg_ordenado_drop_russia.head(15))
plt.savefig('grafico_sem_russia.png')
```

```python
plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Austrália'])
plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Polônia'])
plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Suécia'])
plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Canadá'])
#plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Holanda'])
plota_kg(df_vinho_kg_ordenado_drop_russia.loc['Estados Unidos'])
plt.legend(['Austrália','Polônia',"Suécia", 'Canadá', 'Estados Unidos'])
plt.title('Exportações de vinho nacional entre 2006 a 2021 filtrado')
plt.savefig('grafico_paises_selecionados.png')
```
