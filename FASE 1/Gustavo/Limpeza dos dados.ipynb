## Adicionando as bibliotecas que serão utilizadas

## Importando as bases de dados a serem utilizadas

```python
producao = pd.read_excel("./RawData/Produção_Mundial.xlsx")
producao.head()
```

```python
populacao = pd.read_csv("./RawData/Populacao_Total", sep=";")
populacao.head()
```

Vamos começar fazendo a limpeza das bases "consumo" e "importacao" pela similaridade dos dataframes. 

Observando as bases podemos ver que temos algumas informações que são irrelevantes ou que não fazem sentido estarem no dataset, como por exemplo "Continent", "Product" (Já que todos os dados são referentes a vinhos), "Variable" e "Unit" (Já que todos de quantidade estão na mesma unidade). Por isso vamos retirar essas colunas.

```python
producao.drop(["Continent", "Product", "Unit", "Variable"], axis=1, inplace=True)
producao.head()
```

```python
consumo.drop(["Continent", "Product", "Unit", "Variable"], axis=1, inplace=True)
consumo.head()
```

```python
importacao.drop(["Continent", "Product", "Unit", "Variable"], axis=1, inplace=True)
importacao.head()
```

```python
consumo.tail()
```

```python
importacao.tail()
```

```python
producao.tail() 
```

A seguir vamos aplicar as seguintes melhorias aos nossos datasets para facilitar a manipulação dos dados:

- Traduzir os nomes das colunas e remover caracteres especiais.
- Remover as duas ultimas linhas pois são legendas que vieram do site
- Formatar os dados da coluna Ano

```python
consumo.columns = ["Pais", "Ano", "Quantidade"]
importacao.columns = ["Pais", "Ano", "Quantidade"]
producao.columns = ["Pais", "Ano", "Quantidade"]
```

```python
producao.head()
```

```python
consumo.head()
```

```python
importacao.head()
```

```python
consumo.drop(consumo.tail(2).index, inplace=True)
importacao.drop(importacao.tail(2).index, inplace=True)
producao.drop(producao.tail(2).index, inplace=True)
```

```python
producao.tail()
```

```python
consumo.tail()
```

```python
importacao.tail()
```

```python
producao["Ano"] = producao["Ano"].map(lambda x: int(x))
producao.head()
```

```python
consumo["Ano"] = consumo["Ano"].map(lambda x: int(x))
consumo.head()
```

```python
importacao["Ano"] = importacao["Ano"].map(lambda x: int(x))
importacao.head()
```

Em seguida vamos converter os dados da coluna "Quantidade" para L (Litros) pois no momento se encontram em Hl (Hectolitros).

```python
importacao["Quantidade"] = importacao["Quantidade"].map(lambda x: x*100)
consumo["Quantidade"] = consumo["Quantidade"].map(lambda x: x*100)
producao["Quantidade"] = producao["Quantidade"].map(lambda x: x*100)
```

```python
producao
```

```python
consumo
```

```python
importacao
```

Um outro ajuste que faremos é quanto a data dos dados. Utilizaremos apenas o range de 15 anos para analisar. Portanto filtraremos os dados para que permaneçam apenas os dados entre 2006 e 2021. 2022 Não foi escolhido pois não temos essa informação para a maioria dos países e deixá-lo provocaria algum tipo de alteração em nossa análisa a favor ou contra os países que possuem dados desse ano.

```python
consumo = consumo.query("Ano <= 2021 & Ano >= 2006")
importacao = importacao.query("Ano <= 2021 & Ano >= 2006")
producao = producao.query("Ano <= 2021 & Ano >= 2006")
```

```python
producao
```

```python
consumo
```

```python
importacao
```

Finalizada a limpeza das bases "importacao" e "consumo", podemos olhar agora para nossa base "populacao".

```python
populacao.head()
```

Nela podemos observar que a coluna "Posicao" não nos agrega muito já que a lista está ordenada.

```python
populacao.info()
```

Além disso podemos ver que o dado "Habitantes" está como object. Portanto essas serão os dois ajustes que faremos nessa base.

- Remover a coluna "Posicao"
- Converter os dados da coluna "Habitantes" para inteiro

```python
populacao.drop("Posicao", axis=1, inplace=True)
```

```python
populacao.head()
```

```python
populacao["Habitantes"] = populacao["Habitantes"].str.replace(".","")
populacao["Habitantes"] = populacao["Habitantes"].map(lambda x: int(x))
populacao.head()
```

```python
populacao.info()
```

Assim finalizamos o tratamento de nossas bases.

```python
populacao.to_csv("./TreatedData/Populacao_Total.csv", sep=";", index=False)
importacao.to_csv("./TreatedData/Importacao_Mundial.csv", sep=";", index=False)
consumo.to_csv("./TreatedData/Consumo_Mundial.csv", sep=";", index=False)
producao.to_csv("./TreatedData/Produção_Mundial.csv", sep=";", index=False)
```
