LIMPEZA E MIGRAÇÃO DOS DADOS CLIMATICOS

```python
import pandas as pd
import openpyxl
import os
import csv
import matplotlib.pyplot as plt
```

Leitura dos arquivos RAW e junção para mesmo arquivo

```python
folder_path = r"E:\climatico"

dadosfull = pd.DataFrame()

for filename in os.listdir(folder_path):
    file = "E:/climatico/"+filename
    dados = pd.read_csv(file,skiprows=10,sep=";")
    dadosfull = dadosfull.append(dados, ignore_index=True)
    
```

Tratamento e analise inicial para remoção de colunas extras e dados nulos

```python
dadosfull = dadosfull[dadosfull.columns[:-1]]
dadosfull.shape
```

```python
dadosfull.head()
```

```python
dadosfull.isna()
```

```python
#dadosfull.to_csv('dadosclimaticos.csv',index=False,sep=";")
```

```python
dadosfull1 = dadosfull.dropna()
```

```python
dadosfull1.shape
```

```python
#dadosfull1.to_csv('dadosclimaticos1.csv',index=False,sep=";")
dadosfull1.columns
```

Criação de segundo dataframe temporario para preparação de dados para dataframe final

```python
dadosfull2 = dadosfull1[['Data Medicao', 'PRECIPITACAO TOTAL, DIARIO (AUT)(mm)',
       'TEMPERATURA MAXIMA, DIARIA (AUT)(°C)',
       'TEMPERATURA MINIMA, DIARIA (AUT)(°C)','UMIDADE RELATIVA DO AR, MEDIA DIARIA (AUT)(%)']]

dadosfull2.head()
```

```python
dadosfull2.columns = ['data','precipitacao_total','temp_max','temp_min','humid_media']
dadosfull2
```

```python
dadosfull2['temp_media'] = ((dadosfull2['temp_max'] + dadosfull2['temp_min'])/2)
dadosfull2
dadosfull2.to_csv('dadosfull2.csv',index=False,sep=";",decimal=",")
```

Criação do dataframe final onde será feito um lookup nas datas individuais para cálculo de valores de média

```python
climaticdata_cons = pd.DataFrame(dadosfull2['data'])
climaticdata_cons.head()
```

```python
climaticdata_cons.sort_values('data')
climaticdata_cons.drop_duplicates(inplace=True)
#climaticdata_cons.to_csv('climaticdata_cons.csv',index=False,sep=";")
climaticdata_cons['data'] = pd.to_datetime(climaticdata_cons['data'])
climaticdata_cons.dtypes
```

```python
climatic_data = dadosfull2.groupby('data')
climatic_data_final = climatic_data.mean()
```

```python
#climatic_data_final = climatic_data_final.drop(['temp_max','temp_min'],axis=1)
climatic_data_final = climatic_data_final.reset_index()
climatic_data_final.to_csv('climatic_data_final.csv',index=True,sep=";")
```

```python
climatic_data_final['data'] = pd.to_datetime(climatic_data_final['data'])
climatic_data_final.dtypes
```

```python
climatic_data_final_month = climatic_data_final.groupby(pd.Grouper(key='data', freq='M')).mean()
```

```python
climatic_data_final = climatic_data_final.round(2)
climatic_data_final.to_csv('climatic_data_final.csv',index=True,sep=";",decimal=',')
```

```python
climatic_data_final_month = climatic_data_final_month.round(2)
climatic_data_final_month.to_csv('climatic_data_final_month.csv',index=True,sep=";",decimal=",")
```

```python
climatic_data_final_year = climatic_data_final.groupby(pd.Grouper(key='data', freq='Y')).mean()
climatic_data_final_year = climatic_data_final_year.round(2)
climatic_data_final_year.to_csv('climatic_data_final_year.csv',index=True,sep=";",decimal=",")
```

```python
climatic_data_final_year = climatic_data_final_year.reset_index(drop=False)
climatic_data_final_year
```
