```python
import pandas as pd
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
#climatico
file1=r"E:\analysis\linearegress_clima.csv"


linear_clima = pd.read_csv(file1,sep=';',decimal=',')
```

```python
linear_corr = linear_clima[['precipitacao_total', 'temp_max','humid_media', 'temp_min','Tinto']]
plt.figure(figsize=(10,8))
sns.heatmap(linear_corr.corr(), annot = True, cmap= "RdYlGn");
plt.title('Correlação de Pearson do clima com o Produção de uvas Tintas',size=15);
```

```python
linear_corr = linear_clima[['precipitacao_total', 'temp_max','humid_media', 'temp_min','Brancas/Rosadas']]
plt.figure(figsize=(10,8))
sns.heatmap(linear_corr.corr(), annot = True, cmap= "RdYlGn");
plt.title('Correlação de Pearson do clima com o Produção de uvas Brancas/Rosadas',size=15);
```

```python
linear_corr = linear_clima[['precipitacao_total', 'temp_max','humid_media', 'temp_min','Brancas/Rosadas']]
plt.figure(figsize=(10,8))
plt.title('Correlação de Spearman para uvas Brancas',size=15)
sns.heatmap(linear_corr.corr('spearman'), annot = True, cmap= "RdYlGn");
```

```python
linear_corr = linear_clima[['precipitacao_total', 'temp_max','humid_media', 'temp_min','Tinto']]
plt.figure(figsize=(10,8))
plt.title('Correlação de Spearman para uvas Tintas',size=15)
sns.heatmap(linear_corr.corr('spearman'), annot = True, cmap= "RdYlGn");
```

```python
tinto = ['precipitacao_total', 'temp_max', 'temp_min', 'humid_media','Tinto']
branco = ['precipitacao_total', 'temp_max', 'temp_min', 'humid_media','Brancas/Rosadas']

df_selected = linear_clima[branco]

# Separando os dados em variáveis independentes (X) e dependentes (y)
X = df_selected.drop(columns=['Brancas/Rosadas'])  # Variáveis independentes
y = df_selected['Brancas/Rosadas']  # Variável dependente

# Inicializando o modelo de regressão linear
model = LinearRegression()

# Treinando o modelo com os dados
model.fit(X, y)

# Imprimindo os coeficientes (inclinações) e o intercepto da reta de regressão
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)
print("Score:", model.score(X,y))
```
