```python
dadosfilter = dados[['id', 'control', 'cultivar','2006', '2007', '2008', '2009', '2010',
       '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021']]
```

```python
dadosfilter
```

```python
split_index = 69

dftinto = dadosfilter.iloc[:split_index]
dfbranco = dadosfilter.iloc[split_index:]

dftinto = pd.DataFrame(dftinto)
dfbranco = pd.DataFrame(dfbranco)
```

```python
dftinto
```

```python
dfbranco
```

```python
dftinto_transpose = dftinto.T
#dftinto_transpose = dftinto_transpose.reset_index(drop=True)
dftinto_transpose = dftinto_transpose[2:]
dftinto_transpose.columns = dftinto_transpose.iloc[0]
dftinto_transpose = dftinto_transpose[1:]
dftinto_transpose.rename(columns={'TINTAS': 'Total',"Outras1":"Outras"}, inplace=True)
dftinto_transpose = dftinto_transpose.rename_axis('Ano')
#dftinto_transpose.to_csv('dftinto_transpose.csv',index=True,sep=";")
```

```python
dfbranco_transpose = dfbranco.T
#dfbranco_transpose = dfbranco_transpose.reset_index(drop=True)
dfbranco_transpose = dfbranco_transpose[2:]
dfbranco_transpose.columns = dfbranco_transpose.iloc[0]
dfbranco_transpose = dfbranco_transpose[1:]
dfbranco_transpose.rename(columns={'BRANCAS E ROSADAS': 'Total',"Outras(3)":"Outras"}, inplace=True)
dfbranco_transpose = dfbranco_transpose.rename_axis('Ano')
#dfbranco_transpose.to_csv('dfbranco_transpose.csv',index=True,sep=";")
```

```python
#dfbranco_transpose = dfbranco_transpose.drop(columns=['index'])
#dfbranco_transpose.columns
dfbranco_transpose = dfbranco_transpose.reset_index()
dfbranco_melt = pd.melt(dfbranco_transpose,id_vars='Ano',var_name='Vinhos',value_name="Total")
dfbranco_melt
#dfbranco_melt.to_csv('dfbranco_melt.csv',index=False,sep=";")
dfbranco_melt['Tipo']='Brancas/Rosadas'
```

```python
#dftinto_transpose = dftinto_transpose.drop(columns=['index'])
#dftinto_transpose.columns
dftinto_transpose = dftinto_transpose.reset_index()
dftinto_melt = pd.melt(dftinto_transpose,id_vars='Ano',var_name='Vinhos',value_name="Total")
dftinto_melt
dftinto_melt['Tipo']='Tinto'
#dftinto_melt.to_csv('dftinto_melt.csv',index=False,sep=";")
```

```python
processamentototal = pd.concat([dfbranco_melt,dftinto_melt],ignore_index=True)
processamentototal= processamentototal.replace({0: None, np.nan: None})
processamentototal = processamentototal.dropna()
processamentototal = processamentototal[processamentototal.apply(lambda row: (row != 'nd').all(), axis=1)]
processamentototal.to_csv('processamentototal.csv',index=True,sep=";")
```
